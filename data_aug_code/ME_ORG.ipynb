```python
import pandas as pd
from collections import Counter
```

```python
df = pd.read_csv("../data_aug_sources/middle_eastern_organisations.csv", sep = ";", skiprows = 1)
```

```python
print(df.columns)
```

```python
def add_organisation(organisation):
    tokens = organisation.split()  
    tags = ['B-ORG'] + ['I-ORG'] * (len(tokens) - 1)
    return {"tokens": tokens, "tags": tags}
```

```python
# Checking for duplicates in org
unique_orgs = df["org"].drop_duplicates()

# Adding locations to LOC with correct tags
ORG = [add_organisation(org) for org in unique_orgs]
```

# Counting ORG

```python
def count_org(ORG):
    ORG_lengths = []

    for org in ORG:
        tags = org["tags"]
        if tags[0] == 'B-ORG':
            length = 1
            for tag in tags[1:]:
                if tag == 'I-ORG':
                    length += 1
            ORG_lengths.append(length)

    return Counter(ORG_lengths)
```

```python
print(count_org(ORG))
```
