```python
import pandas as pd
from collections import Counter
```

```python
df = pd.read_csv("../data_aug_sources/the-middle-east-cities.csv", sep = ";", skiprows = 1)
```

```python
print(df.columns)
```

```python
def add_location(location):
    tokens = location.split()  
    tags = ['B-LOC'] + ['I-LOC'] * (len(tokens) - 1)
    return {"tokens": tokens, "tags": tags}
```

```python
# Checking for duplicates in city_da
unique_city_da = df["city_da"].drop_duplicates()

# Adding locations to LOC with correct tags
LOC = [add_location(loc) for loc in unique_city_da]
```

# Counting LOC types/lengths

```python
def count_loc(LOC):
    LOC_lengths = []

    for loc in LOC:
        tags = loc["tags"]
        if tags[0] == 'B-LOC':
            length = 1
            for tag in tags[1:]:
                if tag == 'I-LOC':
                    length += 1
            LOC_lengths.append(length)

    return Counter(LOC_lengths)
```

```python
print(count_loc(LOC))
```
